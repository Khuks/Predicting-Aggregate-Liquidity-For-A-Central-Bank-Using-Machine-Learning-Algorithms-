## Fill Null Values

We will compare two methods of filling null values, namely the mean imputation and interpolate method.


##   Mean Imputation

```python
# We use the mean to fill null values
data=dataset.fillna(dataset.mean())
data
```

#### 1. Dowload The Newly Created Dataset

```python
# We dowload the new dataset that was created after applying mean
data.to_excel('fixed_null_values.xlsx',index=False)
```

#### 2. Check Descriptive Statistics or Summary Statistics

```python
# We get the description of the newly created dataset
data.describe()
```

## Interpolation Method

```python
# We first use the linear method for the interpolate function. The direction for interpolation is the forward direction
forward_data=dataset.fillna(dataset.interpolate(method='linear',limit_direction='forward'))

# Since there were still null values in the first rows of the dataset after applying the forward direction on the interpolate function, 
#  we then apply the backward direction on the newly cleaned data
backward_data=forward_data.fillna(forward_data.interpolate(method='linear',limit_direction='backward'))
```

#### 1. Print the new dataset that was created after using interpolate function

```python
# Download the new dataset
new_data=backward_data.to_excel("interpolated_data.xlsx")
```

#### 2. Get the descriptive statistics of the data after applying the interpolate function

```python
# Get the statistics from the newly created dataset after applying the interpolate function
backward_data.describe()
```

To fill null values, we will choose the interpolate function with linear method. Interpolation is a technique in Python with which you can estimate unknown data points between two known data points. It is commonly used to fill missing values in a table or a dataset using the already known values.

The problem with mean imputation is that the mean imputation does not preserve the relationships among variables. 

Another problem with mean imputation is that it leads to an understimate of standard errors.
